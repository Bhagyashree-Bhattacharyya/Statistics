***Univarariate - Graphical***

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
language = ['Python','R','Java','HTML', 'C++']
popularity = [22.8, 16, 10, 8, 9]
```

```python
colors = ['#FA8F0A','#F0DA32','#61F527','#18DE9D','#3594FF']
```

```python
# PieChart
plt.pie(popularity, labels=language, colors=colors)
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
# Display the % scores in the PieChart
plt.pie(popularity,labels=language, colors=  colors, autopct = "%1.1f%%")
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
# explode a partocular value
explode = (0.2, 0,0,0,0)
plt.pie(popularity,labels=language, colors=  colors, autopct = "%1.1f%%", explode = explode)
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
# Histogram
values = [82,76,24,67,40,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52,65,85]
plt.hist(values)
plt.show()
```

```python
# a modified version
plt.hist(values, bins=10, edgecolor='k', label ="test score data")
plt.title("Histogram for scores")
plt.xlabel("scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Bar Chart
plt.hist(values, bins=10, edgecolor='k', label ="test score data", orientation='horizontal')
plt.title("Histogram for scores")
plt.xlabel("scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# line plot
year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_china = [667.07, 818.32, 981.24, 1135.20, 1262.60, 1337.70]
pop_india = [449.48, 553.58, 696.78, 870.13, 1053.10, 1231.00]
pop_usa =  [180.67, 205.05, 227.23, 249.62, 282.162, 309.34]
```

```python
plt.plot(year,pop_china,label='China',color='y')
plt.plot(year,pop_india,label='India',color='b')
plt.plot(year,pop_usa,label='USA',color='k')
plt.xlabel("Country Names")
plt.ylabel("Pop data")
plt.legend()
plt.show()
```

```python
bank = pd.read_csv('bank.csv', sep = ';')
bank.head()
```

```python
# Grouping with Occupation & subscription profile
bank.groupby('job')['y'].value_counts()
```

```python
# Cross-tabulation
pd.crosstab(bank['job'],bank['y'])
```

```python
# Row-wise normalization
pd.crosstab(bank['job'],bank['y'], normalize='index')
```
