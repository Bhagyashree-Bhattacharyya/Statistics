```python
# What is the probability that a  height of a student is greater than 175 cm and less than 190cm ?

z1 = (175 - df['height'].mean())/df['height'].std()

z2 = (190 - df['height'].mean())/df['height'].std()

from scipy import stats

p = stats.norm.cdf(z2) - stats.norm.cdf(z1)
print(f"probability that a  height of a student is greater than 175 cm and less than 190cm is {p}")
```
