***Transformation of data***

```python
def normality(data, feature):
  plt.figure(figsize=(10,5))
  plt.subplot(1, 2, 1)
  sns.kdeplot(data[feature])
  plt.subplot(1, 2, 2)
  stats.probplot(data[feature], plot = pylab)
  plt.show()
```

```python
normality(cp, "price")
```

Log Transformation

```python
cp['log_price'] = np.log(cp['price'])
normality(cp, "log_price")
```

Reciprocal Transformation

```python
cp['price_reciprocal'] = 1/(cp['price'])
normality(cp, "price_reciprocal")
```

Square Root Transformation

```python
cp['sqrt_price'] = np.sqrt(cp['price'])
normality(cp, "sqrt_price")
```

```python
cp['exponential_price'] = cp['price'] ** (1/1.2)
normality(cp, "exponential_price")
```

Box-Cox transformation

```python
cp['boxcox_price'], parameters = stats.boxcox(cp['price'])
normality(cp, 'boxcox_price')
```
